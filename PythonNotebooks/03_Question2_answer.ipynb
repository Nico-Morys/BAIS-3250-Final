### The purpose of this code is to answer our second question.

#### Question 2. Does a director appear more than once who has highly rated movies and highly sold movies?

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
#reads in the merged csv file with data from both scrapers to answer our question
final_df = pd.read_csv('merged_df.csv', encoding = 'utf-8')

# Sort final_df by gross_earnings in descending order
final_df = final_df.sort_values(by="gross_earnings", ascending=False)

# Count the occurrences of each director in the 'director(s)' column
director_counts = final_df['director(s)'].value_counts()

# Get the top 5 directors with the highest counts
top_3_directors = director_counts.head(3)

# Create a bar plot
plt.figure(figsize=(10, 6))
top_3_directors.plot(kind='bar', color='skyblue')

# Add labels and title
plt.title('Top 3 Directors by Highest Grossing and Highest Rated Movies of All Time')
plt.xlabel('Directors')
plt.ylabel('Director Appearances')
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Calculate the mean of the director counts
mean_director_counts = director_counts.mean()

print(f"The mean number of occurrences per director is: {mean_director_counts}")
```
