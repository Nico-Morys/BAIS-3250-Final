### The purpose of this code is to answer our first question.

#### Question 1. Does movie ranking / rating have anything to do with how much a movie makes?

```python
#imports the libraries we used
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
#reads in the merged csv file with data from both scrapers to answer our question
final_df = pd.read_csv('merged_df.csv', encoding = 'utf-8')

#scatterplot size and x and y axis
plt.figure(figsize=(12, 8))  # Larger figure size for better visualization
plt.scatter(final_df['rank'], final_df['gross_earnings'], color='blue', alpha=0.7)

#set axis limits (we multiplied by 1.2 to make it a bit larger)
plt.xlim(-100, 1000 * 1.2)
plt.ylim(0, 1.5e9 * 1.1)

#format y-axis ticks for readability
formatter = ticker.FuncFormatter(lambda x, pos: f'${x/1e9:.1f}B' if x >= 1e9 else f'${x/1e6:.1f}M')
plt.gca().yaxis.set_major_formatter(formatter)

#add labels and title
plt.title('Scatterplot of Rank vs Gross Earnings', fontsize=16)
plt.xlabel('Rank', fontsize=12)
plt.ylabel('Gross Earnings', fontsize=12)

#adds the grid
plt.grid(True, linestyle='--', alpha=0.6)

#shows the plot
plt.show()
```

```python
#calculates the Pearson correlation coefficient
correlation = final_df['rank'].corr(final_df['gross_earnings'], method='pearson')

print(f"The correlation is: {correlation:.4f}")
```
